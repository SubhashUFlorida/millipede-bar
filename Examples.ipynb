# Example Usage

## Import libraries

```python
from pathlib import Path

import toml

import millipede_bar as mb
```

## Experimental parameters

Load from TOML file 

```python
experimental_parameters = toml.load("parameters.toml")
experimental_parameters
```

Or alternatively, supply as a dictionary

```python
experimental_parameters = {
    "material_properties": {"elastic_modulus": 200.0e9, "density": 7850},
    "geometric_parameters": {"junction_length": 4e-3, "gage_distance": 0},
}
```

## 1D model

### Experimental signal

```python
df_exp = mb.get_incident_data(file=Path("data/experimental.csv"))
df_ana = mb.model_1d(df_incident=df_exp, experimental_parameters=experimental_parameters)
df_ana.plot()
```
